# Predictive Analytics on total cases

```python
# importing necessary packages
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.arima.model import ARIMA
```

```python
# Ensuring the 'Date' column is in datetime format and setting it as the index
aggregated_data['Date'] = pd.to_datetime(aggregated_data['Date'])
aggregated_data.set_index('Date', inplace=True)
aggregated_data.head()
```

```python
# Ensure the original data has a DateTime index
aggregated_data.index = pd.to_datetime(aggregated_data.index)

# Set the frequency of the index (assuming daily data here)
aggregated_data = aggregated_data.asfreq('D')

# Differencing the series to make it stationary
data_diff = aggregated_data['Total Cases'].diff().dropna()
data_diff
```

```python
# ARIMA model fitting on the differenced data
model = ARIMA(data_diff, order=(1, 1, 1))
model_fit = model.fit()

# Forecasting the total cases for the next 30 days and the next 3 months
forecast_months = 90

forecast_months_result = model_fit.get_forecast(steps=forecast_months)

forecast_months_values = forecast_months_result.predicted_mean

# Reverting the differencing to get the forecast in the original scale
forecast_months_values_cumulative = forecast_months_values.cumsum() + aggregated_data['Total Cases'].iloc[-1]

# Plotting the forecasts
forecast_months_dates = pd.date_range(start=aggregated_data.index[-1], periods=forecast_months + 1, inclusive='right')

plt.figure(figsize=(14, 7))
plt.plot(aggregated_data.index, aggregated_data['Total Cases'], label='Historical Data')
plt.plot(forecast_months_dates, forecast_months_values_cumulative, label='3 Month Forecast', color='red')
plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.title('Forecast of Total Cases')
plt.legend()
plt.show()
```

We first note that the y-axis is scaled to scientifc notation 1e9, which means one billion. This shows that we are handling large numbers, which matched our scenario.
